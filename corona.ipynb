```python
data = {
    'Country/Region': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'Confirmed': [36263, 4880, 27973, 907, 950],
    'Recovered': [25198, 2745, 18837, 803, 242]
}

df = pd.DataFrame(data)

# Sort the DataFrame by 'Confirmed' in descending order and select the top 20
df_sorted = df.sort_values(by='Confirmed', ascending=False).head(20)

# Plotting
plt.figure(figsize=(5, 8))
plt.bar(df_sorted['Country/Region'], df_sorted['Confirmed'], label='Confirmed', color='red', alpha=0.7)
plt.bar(df_sorted['Country/Region'], df_sorted['Recovered'], label='Recovered', color='green', alpha=0.7)

plt.xlabel('Country/Region', fontsize=10)
plt.ylabel('Number of Cases', fontsize=10)
plt.title('Top 20 Countries by Confirmed Cases', fontsize=12)
plt.legend()

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```

```python
df_sorted = df.sort_values(by='Confirmed', ascending=True).head(10)

# Plotting total confirmed cases for the top 10 countries
plt.figure(figsize=(8, 3))
plt.barh(df_sorted['Country/Region'], df_sorted['Confirmed'], color='blue', alpha=0.7)

plt.xlabel('Number of Confirmed Cases', fontsize=10)
plt.ylabel('Country/Region', fontsize=10)
plt.title('confirmed cases by top 10 countries', fontsize=12)

plt.tight_layout()

plt.show()
```

```python


df = pd.DataFrame(data)

# Calculate death rate (deaths / confirmed) and sort the DataFrame by it
df['Death Rate'] = df['Deaths'] / df['Confirmed']
df_sorted = df.sort_values(by='Death Rate', ascending=False)

# Plotting death rate for each country
plt.figure(figsize=(8, 6))
plt.barh(df_sorted['Country/Region'], df_sorted['Death Rate'], color='purple', alpha=0.7)

plt.xlabel('Death Rate', fontsize=10)
plt.ylabel('Country/Region', fontsize=10)
plt.title('Death Rate by Country', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
data = {
    'Country/Region': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'Confirmed': [36263, 4880, 27973, 907, 950],
    'Deaths': [1269, 144, 1163, 52, 41],
    'Recovered': [25198, 2745, 18837, 803, 242],
    'Active': [9796, 1991, 7973, 52, 667],
    'New cases': [106, 117, 616, 10, 18],
    'New deaths': [10, 6, 8, 0, 1],
    'New recovered': [18, 63, 749, 0, 0],
}

df = pd.DataFrame(data)

# Sum up the values for the entire dataset
total_active = df['Active'].sum()
total_new_cases = df['New cases'].sum()
total_new_deaths = df['New deaths'].sum()

# Data for the pie chart
labels = ['Active', 'New Cases', 'New Deaths']
sizes = [total_active, total_new_cases, total_new_deaths]
colors = ['yellow', 'lightblue', 'lightcoral']

# Plotting the pie chart
plt.figure(figsize=(5, 10))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Active, New Cases, and New Deaths Across Countries')

plt.show()
```
